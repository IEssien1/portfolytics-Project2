Check for Missing Values

```python
def summarize_missing_values(dataframe):
    summary = {}
    for name, df in dataframe.items():
        missing_values = df.isnull().sum()
        total_values = df.shape[0]
        missing_percentage = (missing_values/total_values) * 100
        missing_df = pd.DataFrame({
            'missing count': missing_values,
            'missing percentage': missing_percentage
        }).query("`missing count` > 0")
        summary[name] = missing_df
    return summary
```

```python
missing_summary = summarize_missing_values({
    'cld': cld,
    'mql': mql,
    'product': product,
    'order': order
})

for name, df in missing_summary.items():
    print(f'summary for {name}')
    if df.empty:
        print('no missing values')
    else:
        print(f'{df}\n')
```

Check for Duplicates

```python
def summarize_duplicates(*datasets, names = None):
    return pd.DataFrame([{
        'Dataset': name,
        'Total_rows': len(df),
        'Duplicate_rows': df.duplicated().sum()
    } for df,name in zip(datasets,names)])
```

Funnel Conversion Rates

1. What is the overall conversion rate from MQLs to closed deals

```python
total_mql = len(mql_clean)
total_cld = len(cld_clean)
conversion_rate = (total_cld / total_mql)*100

plt.figure(figsize=(12, 6))
plt.pie([conversion_rate, 100 - conversion_rate], 
        labels = ['Converted', 'Not Converted'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors = ['skyblue', 'salmon'])
plt.title('Conversion Rate')
plt.show()
```

2. How does the conversion rate vary by marketing channel?

```python
mql_clean.head()
```

```python
cld_clean.head()
```

```python
merged = pd.merge(mql_clean[['mql_id','origin']], cld_clean[['mql_id']], on='mql_id')  
merged.head() 
```

```python
conversion_by_origin = (merged.groupby('origin').size()/mql_clean['origin'].value_counts())*100
conversion_by_origin.head()

plt.figure()
sns.barplot(x=conversion_by_origin.values, 
            y=conversion_by_origin.index,
            hue=conversion_by_origin.index,
            palette='Blues_d',
            legend=False)
plt.title('Conversion Rate by Marketing Channel')
plt.xlabel('Conversion Rate (%)')
plt.ylabel('Marketing Channel')

for index, value in enumerate(conversion_by_origin):
    plt.text(value, index, f'{value:.1f}%')

plt.show()
```

Time-Based Metrics

1. What is the average time taken to close a deal?

```python
merged_data = pd.merge(cld_clean[['mql_id','won_date']], 
                       mql_clean[['mql_id','first_contact_date']], 
                       on='mql_id')
merged_data['time_to_close'] = (merged_data['won_date'] - merged_data['first_contact_date']).dt.days
average_day_to_close = merged_data['time_to_close'].mean()
average_day_to_close

plt.figure(figsize=(12, 6))
sns.histplot(merged_data['time_to_close'], bins=20, kde=True)
plt.title(f'Deal closing Time Distribution: \n Average: {average_day_to_close:.1f} days')
plt.xlabel('Days to Close Deal')
plt.ylabel('Number of Deals')
plt.show()
```

2. How does Conversion velocity vary by time-to-close segments

```python
bins = [0, 7, 15, 30, 60, 365]
labels = ['0-7 days', '8-15 days', '16-30 days', '31-60 days', '60+ days']
merged_data['time_to_close_category'] = pd.cut(merged_data['time_to_close'], bins=bins, labels=labels)  
merged_data

segment_counts = merged_data['time_to_close_category'].value_counts().sort_index()


plt.figure(figsize=(12, 6))
sns.barplot(x=segment_counts.index, 
            y=segment_counts.values, 
            hue=segment_counts.index)
plt.title('Conversion Velocity by Time-to-close Segments')
plt.xlabel('Time to Close')
plt.ylabel('Number of Deals')

total = len(merged_data)
for index, value in enumerate(segment_counts):
    plt.text(index, value, f'{value/total:.1%}', ha='center', va='bottom')
plt.show()
```
